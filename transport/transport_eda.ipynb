```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from streamlit import columns
```

```python
data = pd.read_excel('transport_data.xlsx')
data
```

```python
cleaned_data = data[['СЪЗДАТЕЛ', 'ВЪЗЛОЖИТЕЛ', 'КУРС_ДАТА', 'КУРС', 'ПРЕВОЗВАЧ', 'ШОФЬОР', 'ВЛЕКАЧ', 'ЦИСТЕРНА', 'КМ', 'Л', '€_ЦЕНА_ОБЩО', '€_ЦЕНА_ЗА_1000', 'ДЕН', 'МЕСЕЦ', 'ГОДИНА']]
cleaned_data = cleaned_data.rename(columns={'СЪЗДАТЕЛ': 'ТЪРГОВЕЦ'})
cleaned_data
```

```python
cleaned_data['courses_points'] = cleaned_data['КУРС'].apply(lambda x: x.split('-'))
cleaned_data['points_count'] = cleaned_data['courses_points'].apply(lambda x: len(x) - 2)
cleaned_data = cleaned_data[['ТЪРГОВЕЦ', 'ВЪЗЛОЖИТЕЛ', 'КУРС_ДАТА', 'КУРС', 'points_count', 'ПРЕВОЗВАЧ', 'ШОФЬОР', 'ВЛЕКАЧ', 'ЦИСТЕРНА', 'КМ', 'Л', '€_ЦЕНА_ОБЩО', '€_ЦЕНА_ЗА_1000', 'ДЕН', 'МЕСЕЦ', 'ГОДИНА']]
cleaned_data = cleaned_data.rename(columns={'points_count': 'БРОЙ_ОБЕКТИ'})
cleaned_data
```

```python
cleaned_data.to_csv('transport_data.csv', index=False)
```
