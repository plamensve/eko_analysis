```python
import pandas as pd
```

```python
# -------------------------
# FILE PATH
# -------------------------
file_path = "data/jan.xlsx"

# -------------------------
# READ ALL SHEETS
# -------------------------
xls = pd.ExcelFile(file_path)

dfs = []

for sheet_name in xls.sheet_names:
    df = pd.read_excel(
        xls,
        sheet_name=sheet_name,
        dtype={"Номер на карта": str}
    )

    # clean колоната
    df["Номер на карта"] = (
        df["Номер на карта"]
        .astype(str)
        .str.strip()
        .str.replace(".0", "", regex=False)
    )

    df["company"] = sheet_name
    dfs.append(df)

# -------------------------
# CONCAT
# -------------------------
final_df = pd.concat(dfs, ignore_index=True)

# -------------------------
# SAVE CSV (САМО ТОВА)
# -------------------------
final_df.to_csv(
    "data/jan_comb.csv",
    index=False,
    encoding="utf-8-sig"
)

print("DONE CSV")
```
